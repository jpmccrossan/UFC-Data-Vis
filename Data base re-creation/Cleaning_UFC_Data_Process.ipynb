```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Fighter_Data.csv')
df2 = pd.read_excel('Top_15_and_champion_per_weight_class.xlsx')
```

```python
df = df.rename(columns={'name':'Name', 'wins':'Wins', 'losses':'Losses', 'height':'Height', 'weight':'Weight', 'reach':'Reach', 'stance':'Stance', 'age':'Age',
       'SLpM':'Strikes_Landed_Per_Minute', 'sig_str_acc':'Significant_Strike_Accuracy', 'SApM':'Strikes_Absorbed_Per_Minute', 'str_def':'Strikes_Defended', 'td_avg':'Takedown_Average', 'td_acc':'Takedown_Accuracy', 'td_def':'Takedowns_Defended',
       'sub_avg':'Submission_Average'})
```

```python
df
```

```python
# FIND MISSING NAMES

names_in_df = set(df['Name'])

# Create a set of names in df2
names_in_df2 = set(df2['Name'])

# Check if all names in df2 are present in df
names_missing_from_df_that_are_in_df = names_in_df2 - names_in_df

print(names_missing_from_df_that_are_in_df)
```

```python
row_nums = [51, 230, 1122, 679, 1549, 1021]  

# Create a set of names from df2 and row_nums
names_in_df2 = set(df2['Name'])
names_in_row_nums = set(df.iloc[row_nums]['Name'])

# Combine the sets of names from df2 and row_nums
valid_names = names_in_df2.union(names_in_row_nums)

# Create a filter to keep only the rows with names found in valid_names
filter_condition = df['Name'].isin(valid_names)

# Apply the filter to df
df_filtered = df[filter_condition]

# Display the filtered DataFrame df_filtered
df_filtered

```

```python
df = df_filtered
df
```

```python
df4 = pd.read_excel('Top_15_and_champion_per_weight_class_no_name.xlsx')
df4
```

```python

# Note both df and df4 are alphebetised and correspond 1 for 1
# Assuming df and df4 are the DataFrames
# Resetting the index to ensure sequential index
df.reset_index(drop=True, inplace=True)
df4.reset_index(drop=True, inplace=True)

# Concatenate df and df4 horizontally
concatenated_df = pd.concat([df, df4], axis=1)

concatenated_df
```

```python
# Save concatenated DataFrame to a CSV file
concatenated_df.to_csv('concatenated_styles.csv', index=False)

print("CSV file 'concatenated_styles.csv' has been created.")
```
